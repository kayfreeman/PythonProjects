## üßπ Step 4 ‚Äî Cleaning Column Names
‚úîÔ∏è Purpose
From our investigation above, we need to use consistent column names avoids errors later in calculations and plotting.

- Clean column names

- Fix numeric and date types

- Handle missing values and placeholders

- Start basic aggregation and visualization

- Explore patterns by CUSTOMER, MONTH, SKU

- Make column names consistent and easy to use in code

- Remove spaces, uppercase letters, and special characters

- Use snake_case (lowercase with underscores) ‚Äî Python-friendly

üîπ Teaching Notes

Columns like GROSS AMT ‚Üí gross_amt

PCS ‚Üí pcs

CUSTOMER ‚Üí customer

```python
# Step 4a: Check current column names
print("Original columns:", df.columns.tolist())

# Step 4b: Clean column names
df.columns = (
    df.columns
    .str.strip()           # remove leading/trailing spaces
    .str.lower()           # lowercase all letters
    .str.replace(" ", "_") # replace spaces with underscores
)

# Step 4c: Verify new column names
print("Cleaned columns:", df.columns.tolist())
```

## üîß Step 5 ‚Äî Fixing Data Types & Handling Missing Values
‚úîÔ∏è Purpose

- Convert numeric columns stored as objects to proper numeric types

- Ensure date column is datetime

- Handle missing values in a safe and consistent way

üîπ Teaching Notes

- pcs, rate, gross_amt should be float for calculations.

- Missing numeric values can be filled with 0 or dropped depending on context.

- customer, style, sku may have missing entries ‚Äî can drop rows if they are essential for analysis.

- date must be datetime to allow grouping by month/year

```python
# Step 5a: Convert numeric columns
numeric_cols = ["pcs", "rate", "gross_amt"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")  # invalid entries become NaN

# Step 5b: Convert date column
df["date"] = pd.to_datetime(df["date"], format="%d/%m/%Y", errors="coerce")

# Step 5c: Check for missing values
print("Missing values per column:\n", df.isna().sum())

# Step 5d: Handle missing values
# Example: Drop rows with missing essential info (customer, style, sku)
df = df.dropna(subset=["customer", "style", "sku"])

# Optional: Fill numeric NaNs with 0
df[numeric_cols] = df[numeric_cols].fillna(0)

# Verify data types and missing values
print("\nData types after conversion:\n", df.dtypes)
print("\nShape after cleaning:", df.shape)
```

## üîπ What This Does

- Converts pcs, rate, gross_amt ‚Üí numeric floats

- Converts date ‚Üí datetime

- Drops rows missing essential identifiers

- Fills numeric missing values with 0 to prevent calculation errors

1Ô∏è‚É£ Observed Impact

Rows dropped:

- Original shape: (37432, 10)

- After dropping rows missing customer, style, or sku: (34958, 10)

- 3,474 rows removed (~9.3% of the dataset)

## Implication:

- We‚Äôve removed incomplete rows that cannot be reliably analyzed.

- These rows included placeholder or shipping entries, likely non-product sales.

- Numeric columns converted successfully:

- pcs, rate, gross_amt ‚Üí float64

- NaNs replaced with 0

- Calculations like total gross sales, average rate, and total quantity are now possible.

- Any previous string errors (e.g., ‚Äú1.00‚Äù stored as object) won‚Äôt cause calculation failures.

- Date column converted:

- date ‚Üí datetime64[ns]

- All rows now ready for time-based grouping and plotting

- Enables aggregation by month/year, trend analysis, and filtering by date.

- Remaining missing values:

- months ‚Üí 25 missing (small, can fill from date)

- size ‚Üí 1,040 missing (depends on use; could drop or leave)

- Dataset is now mostly clean and ready for analysis.

- Small missing values can be handled with imputation (e.g., infer months from date) or ignored.

## 2Ô∏è‚É£ Key Implications for EDA & Analysis

- Reliable calculations: All numeric columns can now be used safely in sums, averages, or other aggregations.

- Time-series analysis enabled: You can group by date or derive month/year from the date column.

- Reduced noise: Dropping incomplete rows reduces the chance of misleading insights.

- Remaining small gaps: Minor missing values can be handled case-by-case; dataset is mostly clean.

## Step 7: Basic Aggregations

```python
## 7a: Total Gross Sales by Customer
# Group by 'customer' and sum 'gross_amt'
customer_sales = df.groupby("customer")["gross_amt"].sum().sort_values(ascending=False)

# Display top 10 customers
customer_sales.head(10)
```

```python
## 5b: Top 10 SKUs by Gross Sales
# Group by 'sku' and sum 'gross_amt', then get top 10
sku_sales = df.groupby("sku")["gross_amt"].sum().sort_values(ascending=False).head(10)
sku_sales
```

```python
## 5c: Average Gross Sales per Customer
# Group by 'customer' and compute mean 'gross_amt'
avg_gross_by_customer = df.groupby("customer")["gross_amt"].mean().sort_values(ascending=False).head(10)
avg_gross_by_customer
```

```python
# Total gross sales by customer
customer_sales = df.groupby("customer")["gross_amt"].sum().sort_values(ascending=False)

# Display top 10 customers
print(customer_sales.head(10))

top_customers = customer_sales.head(10)
colors = sns.color_palette("Set3", 10)  # 10 distinct colors
```

## Total sales Per Customer - Basic
üîπ Teaching Notes
- Sum gross_amt for each customer
- Sort to see top customers

```python
plt.figure(figsize=(10,6))
plt.bar(top_customers.index, top_customers.values, color=colors)
plt.title("Top 10 Customers by Gross Sales")
plt.ylabel("Total Gross Amount")
plt.xlabel("Customer")
plt.xticks(rotation=45, ha="right")
plt.show()
```

## Total sales Per Customer - Interactive Chart

```python
# Prepare data
top_customers_df = top_customers.reset_index()
top_customers_df.columns = ["customer", "gross_amt"]

# Create interactive bar chart
fig = px.bar(
    top_customers_df,
    x="customer",
    y="gross_amt",
    color="customer",  # different colors per customer
    color_discrete_sequence=px.colors.qualitative.Set3,
    title="Top 10 Customers by Gross Sales",
    labels={"gross_amt": "Total Gross Amount", "customer": "Customer"}
)

# Update layout for readability
fig.update_layout(
    width=1200,             # make chart wider
    height=600,             # increase height if needed
    xaxis_tickangle=-45,    # rotate x-axis labels
    xaxis_title="Customer",
    yaxis_title="Total Gross Amount",
    showlegend=False        # remove legend (colors already indicate customers)
)

# Show interactive plot
fig.show()
```

## Top 5 Customers Contribution to Gross Sales - Basic Chart

```python
# Top 5 customers
top5_customers = customer_sales.head(5)

plt.figure(figsize=(7,7))
plt.pie(
    top5_customers.values,
    labels=top5_customers.index,
    autopct="%1.1f%%",
    colors=sns.color_palette("Set2", 5),
    startangle=140
)
plt.title("Top 5 Customers Contribution to Gross Sales")
plt.show()
```

```python
# Prepare data
top5_df = top5_customers.reset_index()
top5_df.columns = ["customer", "gross_amt"]

# Create interactive pie chart
fig = px.pie(
    top5_df,
    names="customer",
    values="gross_amt",
    color="customer",
    color_discrete_sequence=px.colors.qualitative.Set2,
    title="Top 5 Customers Contribution to Gross Sales"
)

# Update layout to avoid distortion
fig.update_layout(
    width=900,     # wider chart
    height=700,    # taller chart
    margin=dict(l=50, r=50, t=80, b=50)  # add margin around the chart
)

# Optional: pull slices out slightly for emphasis
fig.update_traces(textinfo='percent+label', pull=[0.05]*len(top5_df))

# Show chart
fig.show()
```

## SKU-Wise Total Gross Amount

```python
sku_sales = df.groupby("sku")["gross_amt"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(sku_sales.index, sku_sales.values, color=sns.color_palette("Paired", 10))
plt.title("Top 10 SKUs by Gross Sales")
plt.ylabel("Total Gross Amount")
plt.xlabel("SKU")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Use one of the built-in qualitative palettes
colors = px.colors.qualitative.Safe  # works in almost all versions
# Other options: 'Pastel1', 'Set1', 'Dark2', 'Vivid', 'Bold'

# Prepare data
sku_sales_df = sku_sales.reset_index()
sku_sales_df.columns = ["sku", "gross_amt"]

# Create interactive bar chart
fig = px.bar(
    sku_sales_df,
    x="sku",
    y="gross_amt",
    color="sku",
    color_discrete_sequence=colors,
    title="Top 10 SKUs by Gross Sales",
    labels={"gross_amt": "Total Gross Amount", "sku": "SKU"}
)

fig.update_layout(
    width=1000,
    height=600,
    xaxis_tickangle=-45,
    showlegend=False
)

fig.show()
```

## Average Gross Amount per Customer
üîπ Shows which customers are high-value buyers (average sale per transaction)

```python
avg_gross_by_customer = df.groupby('customer')['gross_amt'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(avg_gross_by_customer.index, avg_gross_by_customer.values, color=sns.color_palette("Set1", 10))
plt.title("Top 10 Customers by Average Gross Amount")
plt.xlabel("Customer")
plt.ylabel("Average Gross Amount")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Prepare data
avg_gross_df = avg_gross_by_customer.reset_index()
avg_gross_df.columns = ["customer", "avg_gross_amt"]

# Create interactive bar chart
fig = px.bar(
    avg_gross_df,
    x="customer",
    y="avg_gross_amt",
    color="customer",
    color_discrete_sequence=px.colors.qualitative.Set1,  # safe palette
    title="Top 10 Customers by Average Gross Amount",
    labels={"avg_gross_amt": "Average Gross Amount", "customer": "Customer"}
)

# Update layout for readability
fig.update_layout(
    width=1000,
    height=600,
    xaxis_tickangle=-45,
    showlegend=False
)

fig.show()
```
